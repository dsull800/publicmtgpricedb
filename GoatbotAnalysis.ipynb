```python
df=pd.DataFrame(rows,columns=["cardname","carddate","price","cardtype"])
df['carddate'] = df['carddate'].astype('datetime64[ns]') 
```

```python
overalllist=[]
for cardname in df["cardname"].unique():
    for cardtype in df.loc[df['cardname'] == cardname]['cardtype'].unique():
        coeflist=[]
        wow=df['cardname'] == cardname
        wowee=df['cardtype'] == cardtype
        regvals=df.loc[np.logical_and(np.logical_and(wow.values,wowee.values),df['carddate']>'2019-10-01')]
        regvals["carddate"]=regvals["carddate"]-regvals.iloc[0]["carddate"]
        regvals["carddate"]=regvals["carddate"].astype('int64')/86400000000000
        for frac in [1-1/2**i for i in range(0,math.floor(math.log(len(regvals),2)))]:
    #     for frac in [0,.5]:
            if len(regvals.iloc[math.floor(frac*len(regvals)):len(regvals)])<7:
                pass
            else:
                regprice=LinearRegression().fit(np.array(regvals.iloc[math.floor(frac*len(regvals)):len(regvals)]["carddate"]).reshape(-1, 1),regvals.iloc[math.floor(frac*len(regvals)):len(regvals)]["price"])
                pricecoef=regprice.coef_[0]
                coeflist.append((pricecoef,len(regvals.iloc[math.floor(frac*len(regvals)):len(regvals)])))
        coeflist.append(cardtype)
        coeflist.append(cardname)
        overalllist.append(coeflist[::-1])
    
# coefdf=pd.DataFrame(overalllist,columns=["cardname","mecoef1","avgcoef1","medcoef.5","avgcoef.5"])
coefdf=pd.DataFrame(overalllist)
```
